```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Step 1: Create Sales Dataset

data = {
    "Month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun"],
    "Electronics": [12000, 15000, 14000, 17000, 16000, 18000],
    "Clothing": [8000, 9000, 8500, 9500, 10000, 11000],
    "Groceries": [10000, 10500, 11000, 11500, 12000, 12500]
}

df = pd.DataFrame(data)
print(df)
```

```python
# Step 2: Line Chart (Time Series)

plt.figure()
plt.plot(df["Month"], df["Electronics"], marker='o', label="Electronics")
plt.plot(df["Month"], df["Clothing"], marker='o', label="Clothing")
plt.plot(df["Month"], df["Groceries"], marker='o', label="Groceries")

plt.xlabel("Month")
plt.ylabel("Sales Amount")
plt.title("Monthly Sales Trend")
plt.legend()

plt.savefig("monthly_sales_line_chart.png")
plt.show()
```

```python
# Step 3: Bar Chart (Categories)

category_totals = {
    "Electronics": df["Electronics"].sum(),
    "Clothing": df["Clothing"].sum(),
    "Groceries": df["Groceries"].sum()
}

plt.figure()
plt.bar(category_totals.keys(), category_totals.values())

plt.xlabel("Category")
plt.ylabel("Total Sales")
plt.title("Total Sales by Category")

plt.savefig("category_sales_bar_chart.png")
plt.show()
```

```python
# Step 4: Pie Chart (Share)

plt.figure()
plt.pie(
    category_totals.values(),
    labels=category_totals.keys(),
    autopct="%1.1f%%"
)

plt.title("Sales Share by Category")

plt.savefig("category_sales_pie_chart.png")
plt.show()
```

```python
# Step 5: Summary
# ------------------------------
print("""
Project Summary:
- Line chart shows monthly sales trends over time.
- Bar chart compares total sales of each category.
- Pie chart represents percentage share of sales.
- Proper labels, legends, and titles are used.
""")
```

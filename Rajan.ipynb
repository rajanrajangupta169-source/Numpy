```python
import numpy as np
import time
```

```python
# 1D Array
arr1 = np.array([10, 20, 30, 40, 50])

# 2D Array
arr2 = np.array([[1, 2, 3],
                 [4, 5, 6]])

print(arr1)
print(arr2)
```

```python
print(arr1[0])        # First element
print(arr1[1:4])      # Slicing

print(arr2[0, 1])     # Row 0, Column 1
```

```python
print(arr1 + 5)
print(arr1 * 2)

print(np.sum(arr1))
print(np.max(arr1))
print(np.min(arr1))
```

```python
print("Mean:", np.mean(arr1))
print("Median:", np.median(arr1))
print("Standard Deviation:", np.std(arr1))
```

```python
print("Column Sum:", np.sum(arr2, axis=0))
print("Row Sum:", np.sum(arr2, axis=1))
```

```python
arr3 = np.arange(1, 13)
reshaped = arr3.reshape(3, 4)

print(reshaped)

# Broadcasting
print(reshaped + 10)
```

```python
np.save("data.npy", reshaped)

loaded_data = np.load("data.npy")
print(loaded_data)
```

```python
size = 1000000

python_list = list(range(size))
numpy_array = np.arange(size)

# Python List
start = time.time()
python_list = [x * 2 for x in python_list]
print("Python List Time:", time.time() - start)

# NumPy Array
start = time.time()
numpy_array = numpy_array * 2
print("NumPy Array Time:", time.time() - start)
```
